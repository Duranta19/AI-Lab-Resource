```python
from numpy import genfromtxt
data_path = '/content/gdrive/MyDrive/AI/KNN Data Set/iris.csv'
my_data = genfromtxt(data_path, delimiter=',')
```

```python
import random
import math

n = my_data.shape[0]
# print(n)
```

```python
import numpy as np

trainSet = []
validationSet = []
testSet = []
print(my_data[0])
np.random.shuffle(my_data)
print(my_data[0])
for S in my_data:
  r = random.uniform(0,1)
  if r>=0 and r <=0.7:
    trainSet.append(S)
  elif r>0.7 and r <= 0.85:
    validationSet.append(S)
  else:
    testSet.append(S)
```

```python
print(len(trainSet)) 
print(len(validationSet))
# print(validationSet) 
print(len(testSet)) 
```

```python
K = [1,3,5,10,15]
vAcc = []
for x in range(len(K)):
  list1 = []
  # count1 = int(0)
  # count2 = int(0)
  corr = int(0)
  ki = K[x]
  print(ki)

  for V in validationSet:
  # print(V[-1])
    for T in trainSet:
      v0 = V[0]
      v1 = V[1]
      v2 = V[2]
      v3 = V[3]

      t0 = T[0]
      t1 = T[1]
      t2 = T[2]
      t3 = T[3]

      dist = math.sqrt((v0-t0)**2+(v1-t1)**2+(v2-t2)**2+(v3-t3)**2)
      list1.append((dist,T.tolist()))

    list_k = []
    list1.sort(key = lambda x : x[0])
    for i in range(ki):
      list_k.append(list1[i])

    for i in range(ki):
      sample = list_k[i][-1][-1]
      # print(list_k[i])
      # print(sample)
      count1 = int(0)
      count2 = int(0)
      if sample == V[-1]:
        count1 +=1
      else:
        count2 += 1
    if(count1 > count2):
      corr +=1




  print(corr)
  totalSample = float(len(validationSet))
  vAccuracy = (corr/totalSample)*100
  print(vAccuracy)
  vAcc.append(vAccuracy)
  print()



```

```python
max = -9999
point = 0
for i in range(len(vAcc)):
  if(vAcc[i] > max):
    max = vAcc[i]
    point = i
print(point)
```

```python
kt = K[point]
list1 = []
count1 = int(0)
count2 = int(0)
corr = int(0)
for V in testSet:
  # print(V[-1])
  for T in trainSet:
    v0 = V[0]
    v1 = V[1]
    v2 = V[2]
    v3 = V[3]

    t0 = T[0]
    t1 = T[1]
    t2 = T[2]
    t3 = T[3]

    dist = math.sqrt((v0-t0)**2+(v1-t1)**2+(v2-t2)**2+(v3-t3)**2)
    list1.append((dist,T.tolist()))

  list_k = []
  list1.sort(key = lambda x : x[0])
  for i in range(kt):
    list_k.append(list1[i])

  for i in range(kt):
    sample = list_k[i][-1][-1]
    # print(list_k[i])
    # print(sample)
    if sample == V[-1]:
      count1 +=1
    else:
      count2 += 1
  # print(count1-count2)
  if(count1 > count2):
    corr +=1

print(corr)
totalSample = int(len(testSet))
tAccuracy = (corr/totalSample)*100
print(tAccuracy)
```

```python
print("Validation Case")
print("--------------------------")
for i in range(len(K)):
  print("| " + str(K[i]) + " | " + str(vAcc[i]) + "           |")
  print("--------------------------")

print()
print("Test Case")
print("--------------------------")
print("|  " + str(K[point]) + " | " + str(tAccuracy))
print("--------------------------")
```
