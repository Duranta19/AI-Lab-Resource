```python
def init():
  list = [2, 1, 5, 0, 8, 4, 10, 0, 20, 10]
  return list
```

```python
def calculateCost(state):
  cost = 0
  for i in range(len(state)):
    for j in range(i + 1, len(state)):
      if state[i] > state[j]:
        cost +=1
  return cost
```

```python
def generateState(curr_state, curr_cost):

  tempState = curr_state.copy()
  tempCost = curr_cost

  newMinState = []
  for i in range(len(tempState) - 1):
    for j in range(i + 1, len(tempState)):
      temp = tempState[i]
      tempState[i] = tempState[j]
      tempState[j] = temp
      # print(tempState)
      newStateCost = calculateCost(tempState)
      if newStateCost < tempCost:
        tempCost = newStateCost
        newMinState = tempState.copy()
      tempState = curr_state.copy()
  if tempCost < curr_cost:
    return newMinState,tempCost
  else:
    return curr_state,None
```

```python
def goalTest(state):
  if calculateCost(state) == 0:
    return True
  else:
    return False
```

```python
state = init()

while not goalTest(state):
  cost = calculateCost(state)
  state, cost = generateState(state,cost)
  # print(state)
  if cost is None:
    print(state)
    break
print(state)

# print(calculateCost(state))
```
