```python
from queue import PriorityQueue 
```

```python
class Graphs:
  def __init__(self, adjList,heuristic):
    self.adjList = adjList
    self.heuristic = heuristic

  def getNeighbor(self, u):
    return adjList[u]
    
  def aStarSearch(self, src, dist):
    pq = PriorityQueue()
    pq.put(src,0)

    parent = {}
    cost1 = {
        src : 0
    }
    parent[src] = 0

    while not pq.empty():
      currentNode = pq.get()

      if currentNode == dist:
        path = []
        while parent[currentNode] != 0:
          path.append(currentNode)
          currentNode = parent[currentNode]
        path.append(src)
        path.reverse()

        print(path)
        print(cost1[dist])
      
      for v,w in self.getNeighbor(currentNode):
        cost2 = cost1[currentNode] + w
        if v not in cost1 or cost2 < cost1[v]:
          parent[v] = currentNode
          cost1[v] = cost2
          hv = cost2 + heuristic[v]
          pq.put(v,hv)
```
