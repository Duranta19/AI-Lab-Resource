```python
from numpy import genfromtxt
data_path = '/content/gdrive/MyDrive/AI/KNN Data Set/diabetes.csv'
my_data = genfromtxt(data_path, delimiter=',')
# print(my_data[0])
```

```python
import numpy as np
import math
import random
```

```python
trainSet = []
validationSet = []
testSet = []
np.random.shuffle(my_data)

for S in my_data:
  r =random.uniform(0,1)
  if r>= 0 and r <=0.7:
    trainSet.append(S)
  elif r>0.7 and r <=0.85:
    validationSet.append(S)
  else:
    testSet.append(S)
# print(len(trainSet))
# print(len(validationSet))
# print(len(testSet))
```

```python
K = [1,3,5,10,15]
vErr = []
```

```python
for i in range(len(K)):
  ki = K[i]
  list1 = []
  vError = 0.0
  for V in validationSet:
    for T in trainSet:
      v0 = V[0]
      v1 = V[1]
      v2 = V[2]
      v3 = V[3]
      v4 = V[4]
      v5 = V[5]
      v6 = V[6]
      v7 = V[7]
      v8 = V[8]
      v9 = V[9]

      t0 = T[0]
      t1 = T[1]
      t2 = T[2]
      t3 = T[3]
      t4 = T[4]
      t5 = T[5]
      t6 = T[6]
      t7 = T[7]
      t8 = T[8]
      t9 = T[9]

      dist = math.sqrt(((v0-t0)**2) + ((v1-t1)**2) + ((v2-t2)**2) + ((v3-t3)**2) + ((v4-t4)**2) + ((v5-t5)**2)+ ((v6-t6)**2) + ((v7-t7)**2) + ((v8-t8)**2) +((v9-t9)**2))
      list1.append((dist,T.tolist()))
    
    list1.sort()
    list_k = []

    for j in range(ki):
      list_k.append(list1[j])
    sum = 0.0
    for j in range(ki):
      sample = list_k[j][-1][-1]
      sum += sample
    avg = sum/float(ki)
    vError += ((avg - V[10])**2)
  
  Mean_Squared_Error = vError/len(validationSet)
  vErr.append(Mean_Squared_Error)

  
```

```python
print(vErr)
min = 9999
point = 0
for i in range(len(vErr)):
  if vErr[i] < min:
    min = vErr[i]
    point = i
```

```python
ki = K[point]
list1 = []
tError = 0.0
for V in testSet:
  # print(V[10])
  for T in trainSet:
    v0 = V[0]
    v1 = V[1]
    v2 = V[2]
    v3 = V[3]
    v4 = V[4]
    v5 = V[5]
    v6 = V[6]
    v7 = V[7]
    v8 = V[8]
    v9 = V[9]

    t0 = T[0]
    t1 = T[1]
    t2 = T[2]
    t3 = T[3]
    t4 = T[4]
    t5 = T[5]
    t6 = T[6]
    t7 = T[7]
    t8 = T[8]
    t9 = T[9]

    dist = math.sqrt(((v0-t0)**2) + ((v1-t1)**2) + ((v2-t2)**2) + ((v3-t3)**2) + ((v4-t4)**2) + ((v5-t5)**2)+ ((v6-t6)**2) + ((v7-t7)**2) + ((v8-t8)**2) +((v9-t9)**2))
    list1.append((dist,T.tolist()))
  
  list1.sort()
  list_k = []

  for j in range(ki):
    list_k.append(list1[j])
  sum = 0.0
  for j in range(ki):
    sample = list_k[j][-1][-1]
    sum += sample
  avg = sum/float(ki)
  # print(sum)
  # print()
  tError += ((avg - V[10])**2)

Mean_Squared_Error_t = tError/len(testSet)
```
